```python
# Your input text
texto = "Religiosas y mitológicas, es redundante."

# Tokenize the text
inputs = tokenizer.encode_plus(texto, return_tensors="tf", add_special_tokens=True, max_length=512, pad_to_max_length=True, truncation=True)

# Perform text classification using the model
outputs = model(inputs)

# Extract logits for the masked tokens
logits = outputs.logits
print(logits)
# Flatten logits and compute average
logits_avg = tf.reduce_mean(logits, axis=1)
print('avg\n',logits_avg)
# Define your single-layer neural network
num_classes = 2  # Change this according to your task
single_layer_nn = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Forward pass through the single-layer neural network
predictions = single_layer_nn(logits_avg)

# You can then train your single-layer neural network using appropriate loss and optimizer
```
